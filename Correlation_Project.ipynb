## Import Python Libraries

```python
import pandas as pd # import appropriate python library
import seaborn as sns
```

## Load Data

```python
df = pd.read_csv('https://raw.githubusercontent.com/codymbaldwin/sample-files/master/real_estate.csv') # load data
```

## Data Exploration

```python
df.shape # calculating dimension
```

```python
df.info() # checking datatypes
```

```python
df.rename(columns ={'No':'sale_number','X1 transaction date':'sale_date','X2 house age':'house_age','X3 distance to the nearest MRT station':'distance_to_station','X4 number of convenience stores':'num_convenience_stores','X5 latitude':'latitude','X6 longitude':'longitude','Y house price of unit area':'home_price_per_unit'},inplace=True) #renaming columns
```

```python
df.head(7) # preview first 7 records
```

```python
df.describe() # generating descriptive stats.
```

## Generate graph

```python
sns.displot(df,x ='home_price_per_unit'); # generate histogram
```

```python
sns.scatterplot(x ="house_age", y="home_price_per_unit",data=df); #generate scatterplot b/w two variables
```

```python
df.drop(columns=['latitude','longitude','sale_date']).corr() # generate correlation matrix

```

```python
corr = df.drop(columns=['latitude','longitude','sale_date']).corr()
sns.heatmap(corr,annot=True); # generate heatmap
```

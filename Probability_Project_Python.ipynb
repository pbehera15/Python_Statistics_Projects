```python
from scipy import stats  # loading library
```

```python
X = stats.poisson(175)  # creating a variable
```

```python
X.cdf(200) # probablity of 200 or fewer claims
```

```python
1 - X.cdf(200) # probablity of 200 or more claims
```

```python
X.pmf(175) # probablity of getting exactly 175 claims
```

```python
Y = stats.binom(300,0.25) # creating variable
```

```python
Y.pmf(70) # probablity of getting exactly 70
```

```python
1 - Y.cdf(80) # probablity of getting more than 80
```

```python
Y.cdf(90) - Y.cdf(74) # probability of getting between 75 and 90
```
